```python
x=df.drop(columns=['price','car_ID','CarName'])
y=df['price']
```

```python
x_e = pd.get_dummies(x, drop_first=True)
```

```python
x_e
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_e, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "Support Vector Regressor": SVR()
}
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
result={}
for name,model in models.items():
    model.fit(x_train_scaled,y_train)
    y_pred=model.predict(x_test_scaled)
    result[name]={
    'MAE':mean_absolute_error(y_test,y_pred),
    'MSE':mean_squared_error(y_test,y_pred),
    'r2':r2_score(y_test,y_pred)
    }
```

```python
result_df=pd.DataFrame(result).T
```

```python
result_df
```

```python
best_model = result_df['r2'].idxmax()
best_model
```

Random forest is best beacuse of its high r2 value

```python
rand=RandomForestRegressor()
rand.fit(x_train_scaled,y_train)
importan_features=rand.feature_importances_ 
```

```python
importan_features
```

```python
e = pd.DataFrame({
    'Feature': x_e.columns,
    'Importance': importan_features
})
```

```python
e = e.sort_values(by='Importance', ascending=False)
```

```python
e.head()
```

These are top 5 features affecting car price

```python
plt.figure(figsize=(10,6))
plt.barh(e['Feature'][:10], e['Importance'][:10])
plt.xlabel("Importance")
plt.title("Top 10 Important Features")
plt.show()
```

```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 15, None],
    'min_samples_split': [2, 5]
}
```
